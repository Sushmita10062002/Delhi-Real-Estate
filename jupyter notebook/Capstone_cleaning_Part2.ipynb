```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("flats_houses_v1.csv")
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.head(1)
```

```python
df.drop(columns = ["PROP_NAME", "BUILDING_NAME", "CAPACITY", "SHARING_COUNT", "AGE"], inplace = True)
```

```python
df.head(2)
```

```python
df.dropna(subset=["AVG_PRICE"], inplace=True)
```

```python
df.info()
```

```python
def treat_price(x):
  x = x/10000000
  return round(x, 2)
```

```python
df["AVG_PRICE_CR"] = df["AVG_PRICE"].apply(treat_price)
```

```python
df.head()
```

```python
df = df[~df["BEDROOM_NUM"].isnull()]
```

```python
# df = df[df["AVAILABILITY"] != "Under Construction"]
```

```python
def floor_num_fun(floor_num):
  if floor_num == 'L':
    return '0'
  elif floor_num == 'B':
    return '-1'
  elif floor_num == 'G':
    return '0'
  elif floor_num == 'M':
    return '3'
  else:
    return floor_num
```

```python
df["FLOOR_NUM"] = df["FLOOR_NUM"].apply(floor_num_fun)
```

```python
df["FLOOR_NUM"].value_counts()
```

```python
df["FACING"].isnull().sum()
```

```python
df["FACING"].fillna("NA", inplace = True)
```

```python
df["AREA_SQFT"] = df["AVG_PRICE"]/df["PRICE_SQFT"]
```

```python
df[["AREA_SQFT", "BUILTUP_SQFT", "CARPET_SQFT", "SUPERBUILTUP_SQFT"]][100:120]
```

```python
def prop_type_fun(property_type):
  if property_type == "Residential Apartment":
    return "flat"
  elif property_type == "Independent House/Villa":
    return "house"

df["PROPERTY_TYPE"] = df["PROPERTY_TYPE"].apply(prop_type_fun)
```

```python
df.to_csv("flat_houses_v2.csv", index = False)
```
